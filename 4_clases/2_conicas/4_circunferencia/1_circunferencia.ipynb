```python
import pandas as pd

# Ruta del archivo
path = 'C:/Users/marco/Documentos/docencia/groups_list/vectorial.xls'

# Leer el archivo como CSV o TSV
# Si el archivo está separado por tabulaciones (TSV)
estudiantes = pd.read_csv(path, sep=',', encoding='ISO-8859-1')
estudiantes.head(4)  # Mostrar las primeras 3 filas
```

```python

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    global lista_estudiantes
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        # Validación de entrada
        if asistencia == 'n':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Asistencia'] += 4
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Asistencia']} en la columna 'Parcial'.")
        
        elif asistencia != 's':
            print("Entrada inválida. Por favor, introduce 's' para sí o 'n' para no.")
            return  # Volver a preguntar para el mismo estudiante si la entrada es inválida
        
        # Remover al estudiante de la lista para no llamarlo de nuevo
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Función para guardar cambios al final de la sesión
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()

# Guardar los cambios una vez que todos los estudiantes hayan sido llamados
guardar_cambios()

```

### **Metodología de Enseñanza: Aprendizaje Basado en Problemas y Proyectos (ABPP)**



El **Aprendizaje Basado en Problemas y Proyectos (ABPP)** es una metodología educativa que integra los enfoques del **Aprendizaje Basado en Problemas (ABP)** y el **Aprendizaje Basado en Proyectos (PBL, por sus siglas en inglés)**. 

Es especialmente útil para la formación de futuros docentes, ingenieros y administradores, ya que combina el desarrollo de habilidades prácticas con el pensamiento crítico y reflexivo. 

A continuación, se presenta una descripción adaptada a la enseñanza de las matemáticas en educación infantil, Licenciatura en Matemáticas, Ingeniería Agropecuaria y Administración de Empresas:

---



### **1. Fundamentos del ABPP**
El ABPP se basa en:
- **Resolución de Problemas Reales:** Los estudiantes analizan y resuelven problemas auténticos relacionados con su área de formación, fomentando la aplicación práctica del conocimiento.


- **Proyectos como Herramientas Integradoras:** Los proyectos permiten desarrollar habilidades interdisciplinarias y ofrecer soluciones concretas a los problemas planteados.


- **Aprendizaje Activo:** Los estudiantes son responsables de su aprendizaje y participan activamente en el proceso de investigación, diseño y solución.

---



### **2. Aplicación del ABPP en Áreas Específicas**



#### **Educación Infantil (Licenciatura en Educación Infantil)**
- **Enfoque:** Desarrollo del pensamiento matemático en la infancia desde contextos lúdicos, familiares y culturales.


- **Problemas típicos:** 
  - ¿Cómo enseñar los números naturales mediante actividades prácticas y juegos?
  - ¿Qué estrategias pueden emplearse para abordar el concepto de cantidad en edades tempranas?


- **Proyectos:** Diseño de secuencias didácticas que usen Python, Jupyter Notebooks y materiales manipulativos para enseñar matemáticas de forma interactiva.


- **Competencias desarrolladas:** 
  - Uso de herramientas tecnológicas para diseñar actividades educativas.
  - Comprensión del desarrollo cognitivo infantil y su relación con las matemáticas.



#### **Licenciatura en Matemáticas**
- **Enfoque:** Resolución de problemas matemáticos aplicados y profundización teórica en áreas como análisis matemático y modelado.


- **Problemas típicos:**
  - ¿Cómo se pueden resolver ecuaciones diferenciales aplicadas a modelos físicos o económicos?
  - ¿Cómo modelar fenómenos naturales mediante funciones matemáticas?


- **Proyectos:** 
  - Desarrollo de software educativo para explorar conceptos matemáticos.
  - Aplicación de modelado matemático para resolver problemas reales en otras disciplinas.


- **Competencias desarrolladas:**
  - Dominio de la matemática teórica y aplicada.
  - Uso de tecnología para la enseñanza y la investigación.



#### **Ingeniería Agropecuaria**


- **Enfoque:** Modelado matemático y análisis de datos para mejorar la producción agrícola y la gestión de recursos.
- **Problemas típicos:**
  - ¿Cómo modelar el crecimiento de cultivos considerando variables climáticas?
  - ¿Cómo optimizar el uso del agua en sistemas de riego mediante análisis matemático?


- **Proyectos:** 
  - Desarrollo de modelos predictivos con Python y bibliotecas de aprendizaje automático.
  - Implementación de proyectos que combinen matemáticas y ciencias de datos para resolver problemas agrícolas.


- **Competencias desarrolladas:**
  - Integración de matemáticas con tecnologías avanzadas.
  - Resolución de problemas agropecuarios mediante análisis y modelado.



#### **Administración de Empresas**
- **Enfoque:** Uso de las matemáticas y la ciencia de datos para la toma de decisiones estratégicas.


- **Problemas típicos:**
  - ¿Cómo identificar patrones de comportamiento del consumidor usando series de tiempo?
  - ¿Qué estrategias matemáticas pueden optimizar la asignación de recursos en una empresa?


- **Proyectos:** 
  - Implementación de modelos de análisis financiero y simulaciones.
  - Desarrollo de dashboards interactivos para la gestión de datos empresariales.


- **Competencias desarrolladas:**
  - Habilidad para interpretar datos matemáticos en contextos de negocio.
  - Capacidad para diseñar soluciones basadas en datos.

---



### **3. Beneficios del ABPP**


- **Desarrollo de competencias clave:**
  - Pensamiento crítico y creativo.
  - Trabajo colaborativo.
  - Comunicación efectiva.
  - Uso de herramientas tecnológicas avanzadas.


- **Adaptación al contexto laboral:** Los estudiantes aprenden a enfrentar desafíos del mundo real y a diseñar soluciones prácticas y aplicables.
- **Flexibilidad e interdisciplinariedad:** La metodología se adapta a diversas áreas, fomentando el aprendizaje integrado.

```python
def trazar_circunferencia_y_cuatro_puntos_cardinales(centro, radio): 
    """_summary_

    Args:
        centro (tuple): tupla o par ordenado con las coordenadas x e y del centro de la circunferencia.
        radio (float): número real que representa el radio de la circunferencia.
    """
    import matplotlib.pyplot as plt
    import numpy as np
    # Representación paramétrica de la circunferencia. 
    theta = np.linspace(0, 2 * np.pi, 100)
    h, k = centro[0], centro[1]
    r = radio
    x = h + r * np.cos(theta)
    y = k + r * np.sin(theta)
    # 1. Creación del lienzo
    plt.figure(figsize=(7, 7))
    # 2. Trazado de la circunferencia
    plt.plot(x, y, label='Circunferencia')
    # 3. trazado del centro
    plt.scatter(h, k, color='red', label=f'Centro ({h}, {k})')
    # 4. Trazado de cuatro puntos cardinales de la circunferencia
    oriente = (h+r, k)
    norte = (h, k+r)
    poniente = (h-r, k) 
    sur = (h, k-r)
    abscisas_puntos_cardinales, ordenadas_puntos_cardinales = zip(oriente, norte, poniente, sur)
    plt.scatter(abscisas_puntos_cardinales, ordenadas_puntos_cardinales, color='green', label='Puntos Cardinales')
    plt.axis('equal')
    plt.legend(loc = 'upper right')
    plt.title(f'Circunferencia $(x - {h})^2 + (y - {k})^2 = {r}^2$')
    plt.grid(alpha = 0.4)
    plt.xlabel('eje de las abscisas x')
    plt.ylabel('eje de las ordenadas y')
    plt.xticks(np.arange(h-r, h+r+1, 1))
    plt.yticks(np.arange(k-r, k+r+1, 1))
    save_fig('circunferencia_centro_radio')
    plt.show()

```

### **Diferencia entre la Ecuación General de Segundo Grado y la Ecuación Canónica de una Circunferencia**



La **circunferencia** es una de las secciones cónicas más estudiadas en geometría analítica. Su representación matemática puede darse en dos formas principales: la **ecuación general de segundo grado** y la **ecuación canónica**. A continuación, te explico en detalle la diferencia entre ambas:

---



### **1. Ecuación Canónica de la Circunferencia**

La **ecuación canónica** de una circunferencia es la forma más simple y directa de representarla. Se expresa como:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

Donde:
- $(h, k)$ es el **centro** de la circunferencia.
- $r$ es el **radio** de la circunferencia.



**Características:**
- Es una forma **reducida** y fácil de interpretar.
- Permite identificar directamente el centro y el radio de la circunferencia.
- Es útil para graficar la circunferencia rápidamente.

**Ejemplo:**  
La ecuación $(x - 2)^2 + (y + 3)^2 = 16$ representa una circunferencia con:
- Centro en $(2, -3)$.
- Radio $r = 4$ (ya que $r^2 = 16$).

---



### **2. Ecuación General de Segundo Grado**

La **ecuación general de segundo grado** para una circunferencia es una forma más amplia y se expresa como:

$$
x^2 + y^2 + Dx + Ey + F = 0
$$

Donde:
- $D$, $E$ y $F$ son coeficientes constantes.
- Esta forma incluye términos lineales ($Dx$ y $Ey$) y un término independiente ($F$).



**Características:**
- Es una forma **expandida** de la ecuación canónica.
- No permite identificar directamente el centro y el radio; se requiere completar cuadrados para convertirla a la forma canónica.
- Es útil en contextos teóricos y para resolver sistemas de ecuaciones.

**Ejemplo:**  
La ecuación $x^2 + y^2 - 4x + 6y - 3 = 0$ representa una circunferencia, pero no es evidente cuál es su centro o radio. Para encontrar estas propiedades, debemos completar cuadrados.

---



### **3. Conversión de la Ecuación General a la Forma Canónica**

Para convertir la ecuación general a la forma canónica, seguimos estos pasos:



1. **Agrupar términos:**  
   Agrupamos los términos en $x$ y los términos en $y$:

   $$
   x^2 + Dx + y^2 + Ey = -F
   $$



2. **Completar cuadrados:**  
   Completamos cuadrados para los términos en $x$ y $y$:

   - Para $x$:
     $$
     x^2 + Dx \quad \Rightarrow \quad \left(x + \frac{D}{2}\right)^2 - \left(\frac{D}{2}\right)^2
     $$

   - Para $y$:
     $$
     y^2 + Ey \quad \Rightarrow \quad \left(y + \frac{E}{2}\right)^2 - \left(\frac{E}{2}\right)^2
     $$



3. **Reescribir la ecuación:**  
   Sustituimos los cuadrados completados en la ecuación:

   $$
   \left(x + \frac{D}{2}\right)^2 + \left(y + \frac{E}{2}\right)^2 = \left(\frac{D}{2}\right)^2 + \left(\frac{E}{2}\right)^2 - F
   $$



4. **Identificar centro y radio:**  
   Comparando con la forma canónica $(x - h)^2 + (y - k)^2 = r^2$, obtenemos:
   - Centro: $(h, k) = \left(-\frac{D}{2}, -\frac{E}{2}\right)$.
   - Radio: $r = \sqrt{\left(\frac{D}{2}\right)^2 + \left(\frac{E}{2}\right)^2 - F}$.



**Ejemplo:**  
Dada la ecuación general $x^2 + y^2 - 4x + 6y - 3 = 0$:
1. Agrupamos términos:
   $$
   x^2 - 4x + y^2 + 6y = 3
   $$

2. Completamos cuadrados:
   - Para $x$:
     $$
     x^2 - 4x \quad \Rightarrow \quad (x - 2)^2 - 4
     $$

   - Para $y$:
     $$
     y^2 + 6y \quad \Rightarrow \quad (y + 3)^2 - 9
     $$

3. Reescribimos la ecuación:
   $$
   (x - 2)^2 - 4 + (y + 3)^2 - 9 = 3
   $$

4. Simplificamos:
   $$
   (x - 2)^2 + (y + 3)^2 = 16
   $$

5. Identificamos centro y radio:
   - Centro: $(2, -3)$.
   - Radio: $r = 4$.

### **4. Resumen de Diferencias**

| **Característica**            | **Ecuación Canónica**                          | **Ecuación General**                          |
|-------------------------------|-----------------------------------------------|-----------------------------------------------|
| **Forma**                     | $(x - h)^2 + (y - k)^2 = r^2$                 | $x^2 + y^2 + Dx + Ey + F = 0$                 |
| **Centro y Radio**            | Directamente identificables                   | Requiere completar cuadrados para identificarlos |
| **Uso**                       | Gráficos y aplicaciones directas              | Contextos teóricos y sistemas de ecuaciones   |
| **Simplicidad**               | Más simple y directa                          | Más compleja, requiere transformación         |

---



### **5. Ejemplo en Python**

Aquí tienes un código en Python para graficar una circunferencia a partir de su ecuación general:


```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes de la ecuación general: x^2 + y^2 + Dx + Ey + F = 0
D, E, F = -4, 6, -3

# Convertir a forma canónica
h = -D / 2
k = -E / 2
r = np.sqrt((D/2)**2 + (E/2)**2 - F)

# Graficar la circunferencia
theta = np.linspace(0, 2 * np.pi, 100)
x = h + r * np.cos(theta)
y = k + r * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Circunferencia')
plt.scatter(h, k, color='red', label=f'Centro ({h}, {k})')
plt.axis('equal')
plt.legend()
plt.title(f'Circunferencia $(x - {h})^2 + (y - {k})^2 = {r**2}$')
plt.show()

```


---

### **Conclusión**

- La **ecuación canónica** es la forma más directa y útil para graficar y analizar una circunferencia, ya que muestra explícitamente su centro y radio.
- La **ecuación general** es más versátil en contextos teóricos, pero requiere un proceso de transformación (completar cuadrados) para identificar sus propiedades geométricas.

## **1. Sistemas de Riego Circulares**



### **Descripción:**
Los sistemas de riego circulares, también conocidos como **pivotes centrales**, son sistemas de riego mecanizado que giran alrededor de un punto central, distribuyendo agua de manera uniforme en un área circular.

### **Ventajas:**
1. **Eficiencia en el uso del agua:**  
   - La distribución uniforme del agua reduce el desperdicio y asegura que todas las plantas reciban la misma cantidad de agua.
   - Ideal para zonas con escasez de agua.

2. **Cobertura amplia:**  
   - Puede regar grandes extensiones de terreno con un solo sistema.

3. **Ahorro de mano de obra:**  
   - Al ser automatizados, requieren menos intervención humana en comparación con los sistemas de riego tradicionales.

4. **Adaptabilidad:**  
   - Puede usarse en terrenos irregulares o con pendientes moderadas.

5. **Aplicaciones:**  
   - Cultivos extensivos como maíz, trigo, soja y pastos.

---



## **2. Cultivos Piscícolas en Tanques Circulares**

### **Descripción:**
Los tanques circulares son estructuras utilizadas en la piscicultura para criar peces. Su forma circular favorece la circulación del agua y la distribución uniforme de los nutrientes.

### **Ventajas:**
1. **Mejor circulación del agua:**  
   - La forma circular permite un flujo continuo y uniforme del agua, lo que mejora la oxigenación y elimina desechos de manera eficiente.

2. **Facilidad de limpieza:**  
   - Los tanques circulares son más fáciles de limpiar y mantener que los tanques rectangulares, ya que no tienen esquinas donde se acumulen desechos.

3. **Reducción de estrés en los peces:**  
   - El flujo circular del agua simula condiciones naturales, reduciendo el estrés en los peces y mejorando su crecimiento.

4. **Optimización del espacio:**  
   - Los tanques circulares permiten una distribución más eficiente del espacio en las granjas piscícolas.

5. **Aplicaciones:**  
   - Cría de tilapia, trucha, camarones y otras especies acuícolas.

---

## **3. Potreros Circulares**

### **Descripción:**
Los potreros circulares son áreas de pastoreo diseñadas en forma circular, generalmente alrededor de un punto central donde se ubican los bebederos o comederos.

### **Ventajas:**
1. **Distribución equitativa de recursos:**  
   - Los animales tienen acceso uniforme al agua y alimento, lo que reduce la competencia y el estrés.

2. **Facilidad de manejo del ganado:**  
   - La forma circular permite un movimiento más natural del ganado, facilitando su manejo y reduciendo lesiones.

3. **Optimización del pastoreo:**  
   - El diseño circular favorece un pastoreo más uniforme, evitando la sobreexplotación de ciertas áreas.

4. **Reducción de la erosión:**  
   - Al evitar esquinas y bordes rectos, se reduce la erosión del suelo y se promueve un uso más sostenible del terreno.

5. **Aplicaciones:**  
   - Ganadería extensiva, especialmente en sistemas de pastoreo rotacional.

### **Ventaja Geométrica de los Potreros Circulares frente a los Potreros Rectangulares del Mismo Perímetro**

Una de las ventajas más importantes de los potreros circulares frente a los rectangulares es su **mayor área útil** para el mismo perímetro. Esta propiedad geométrica se deriva de un principio matemático fundamental: **entre todas las figuras planas con un perímetro dado, el círculo es la que encierra la mayor área**.

---



### **1. Explicación Matemática**

#### **Fórmulas clave:**
- **Perímetro de un círculo:**  
  $$
  P_{\text{círculo}} = 2\pi r
  $$
  Donde $r$ es el radio del círculo.



- **Área de un círculo:**  
  $$
  A_{\text{círculo}} = \pi r^2
  $$

- **Perímetro de un rectángulo:**  
  $$
  P_{\text{rectángulo}} = 2(l + a)
  $$
  Donde $l$ es el largo y \( a \) es el ancho.



- **Área de un rectángulo:**  
  $$
  A_{\text{rectángulo}} = l \cdot a
  $$



#### **Comparación de áreas:**
Supongamos que tenemos un potrero circular y un potrero rectangular con el mismo perímetro $P$.

1. **Para el círculo:**
   - El radio se calcula como:
     $$
     r = \frac{P}{2\pi}
     $$
   - El área es:
     $$
     A_{\text{círculo}} = \pi \left(\frac{P}{2\pi}\right)^2 = \frac{P^2}{4\pi}
     $$

2. **Para el rectángulo:**
   - Supongamos un rectángulo cuadrado (el rectángulo con mayor área para un perímetro dado). En este caso, \( l = a \).
   - El lado se calcula como:
     $$
     l = \frac{P}{4}
     $$
   - El área es:
     $$
     A_{\text{rectángulo}} = \left(\frac{P}{4}\right)^2 = \frac{P^2}{16}
     $$



3. **Comparación:**
   - El área del círculo es:
     $$
     A_{\text{círculo}} = \frac{P^2}{4\pi} \approx \frac{P^2}{12.57}
     $$
   - El área del rectángulo cuadrado es:
     $$
     A_{\text{rectángulo}} = \frac{P^2}{16}
     $$

   Claramente, \( \frac{P^2}{12.57} > \frac{P^2}{16} \), lo que significa que el círculo tiene un área mayor que el rectángulo para el mismo perímetro.

---



### **2. Implicaciones Prácticas**

#### **Mayor área útil:**
- Un potrero circular proporciona más espacio para el pastoreo que un potrero rectangular con el mismo perímetro. Esto es especialmente útil en regiones donde el terreno es limitado o costoso.

#### **Menor longitud de cerco:**
- Para un área dada, un potrero circular requiere menos longitud de cerco que un potrero rectangular. Esto reduce los costos de construcción y mantenimiento del cerco.

#### **Distribución uniforme:**
- La forma circular permite una distribución más uniforme de los recursos (agua, alimento) desde un punto central, lo que reduce la competencia entre los animales y mejora su bienestar.

---



### **3. Ejemplo Numérico**

Supongamos un perímetro $ P = 100$ metros.

1. **Potrero circular:**
   - Radio:
     $$
     r = \frac{100}{2\pi} \approx 15.92 \text{ metros}
     $$
   - Área:
     $$
     A_{\text{círculo}} = \pi (15.92)^2 \approx 796 \text{ m}^2
     $$

2. **Potrero rectangular cuadrado:**
   - Lado:
     $$
     l = \frac{100}{4} = 25 \text{ metros}
     $$
   - Área:
     $$
     A_{\text{rectángulo}} = 25 \times 25 = 625 \text{ m}^2
     $$



3. **Comparación:**
   - El potrero circular tiene un área de **796 m²**, mientras que el potrero rectangular tiene un área de **625 m²**. Esto significa que el potrero circular ofrece un **27% más de área útil** para el mismo perímetro.

---



### **4. Conclusión**

La ventaja geométrica de los potreros circulares radica en su capacidad para maximizar el área útil con un perímetro dado. Esto no solo optimiza el uso del terreno, sino que también reduce costos de construcción y mejora la eficiencia en la distribución de recursos. En aplicaciones agropecuarias, esta propiedad hace que los potreros circulares sean una opción más eficiente y sostenible que los rectangulares.

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2 * np.pi, 100)
x = 2 + 2 * np.cos(theta)
y = 3 + 2 * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Circunferencia')
plt.scatter(2, 3, color='red', label='Centro (2, 3)')
plt.axis('equal')
plt.legend(loc = 'upper right')
plt.title('Circunferencia $(x - 2)^2 + (y - 3)^2 = 4$')
plt.grid(alpha = 0.4)
plt.show()

```


2. **Actividad 2: Simulación de riego circular**  
   - Crear una animación que muestre cómo el agua se distribuye en un sistema de riego circular.  


```python


from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots()
circle = plt.Circle((2, 3), 2, fill=False, color='blue')
ax.add_patch(circle)
ax.set_xlim(0, 4)
ax.set_ylim(0, 6)

def animate(i):
    ax.clear()
    ax.add_patch(circle)
    ax.scatter(2 + 2 * np.cos(i * 0.1), 3 + 2 * np.sin(i * 0.1), color='green')
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 6)

ani = FuncAnimation(fig, animate, frames=100, interval=100)
HTML(ani.to_jshtml())

```
