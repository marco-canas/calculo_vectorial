### **Sistemas de Riego Circulares, Cultivos Piscícolas en Tanques Circulares y Potreros Circulares: Ventajas y Aplicaciones**

La geometría circular es ampliamente utilizada en aplicaciones agropecuarias debido a sus propiedades únicas, como la simetría, la eficiencia en la distribución de recursos y la facilidad de diseño. A continuación, te explico en detalle los sistemas de riego circulares, los cultivos piscícolas en tanques circulares y los potreros circulares, junto con sus ventajas.

---



## **1. Sistemas de Riego Circulares**

### **Descripción:**
Los sistemas de riego circulares, también conocidos como **pivotes centrales**, son sistemas de riego mecanizado que giran alrededor de un punto central, distribuyendo agua de manera uniforme en un área circular.

### **Ventajas:**
1. **Eficiencia en el uso del agua:**  
   - La distribución uniforme del agua reduce el desperdicio y asegura que todas las plantas reciban la misma cantidad de agua.
   - Ideal para zonas con escasez de agua.

2. **Cobertura amplia:**  
   - Puede regar grandes extensiones de terreno con un solo sistema.

3. **Ahorro de mano de obra:**  
   - Al ser automatizados, requieren menos intervención humana en comparación con los sistemas de riego tradicionales.

4. **Adaptabilidad:**  
   - Puede usarse en terrenos irregulares o con pendientes moderadas.

5. **Aplicaciones:**  
   - Cultivos extensivos como maíz, trigo, soja y pastos.

---

## **2. Cultivos Piscícolas en Tanques Circulares**

### **Descripción:**
Los tanques circulares son estructuras utilizadas en la piscicultura para criar peces. Su forma circular favorece la circulación del agua y la distribución uniforme de los nutrientes.

### **Ventajas:**
1. **Mejor circulación del agua:**  
   - La forma circular permite un flujo continuo y uniforme del agua, lo que mejora la oxigenación y elimina desechos de manera eficiente.

2. **Facilidad de limpieza:**  
   - Los tanques circulares son más fáciles de limpiar y mantener que los tanques rectangulares, ya que no tienen esquinas donde se acumulen desechos.

3. **Reducción de estrés en los peces:**  
   - El flujo circular del agua simula condiciones naturales, reduciendo el estrés en los peces y mejorando su crecimiento.

4. **Optimización del espacio:**  
   - Los tanques circulares permiten una distribución más eficiente del espacio en las granjas piscícolas.

5. **Aplicaciones:**  
   - Cría de tilapia, trucha, camarones y otras especies acuícolas.

---

## **3. Potreros Circulares**

### **Descripción:**
Los potreros circulares son áreas de pastoreo diseñadas en forma circular, generalmente alrededor de un punto central donde se ubican los bebederos o comederos.

### **Ventajas:**
1. **Distribución equitativa de recursos:**  
   - Los animales tienen acceso uniforme al agua y alimento, lo que reduce la competencia y el estrés.

2. **Facilidad de manejo del ganado:**  
   - La forma circular permite un movimiento más natural del ganado, facilitando su manejo y reduciendo lesiones.

3. **Optimización del pastoreo:**  
   - El diseño circular favorece un pastoreo más uniforme, evitando la sobreexplotación de ciertas áreas.

4. **Reducción de la erosión:**  
   - Al evitar esquinas y bordes rectos, se reduce la erosión del suelo y se promueve un uso más sostenible del terreno.

5. **Aplicaciones:**  
   - Ganadería extensiva, especialmente en sistemas de pastoreo rotacional.

---

### **Ventaja Geométrica de los Potreros Circulares frente a los Potreros Rectangulares del Mismo Perímetro**

Una de las ventajas más importantes de los potreros circulares frente a los rectangulares es su **mayor área útil** para el mismo perímetro. Esta propiedad geométrica se deriva de un principio matemático fundamental: **entre todas las figuras planas con un perímetro dado, el círculo es la que encierra la mayor área**.

---



### **1. Explicación Matemática**

#### **Fórmulas clave:**
- **Perímetro de un círculo:**  
  $$
  P_{\text{círculo}} = 2\pi r
  $$
  Donde \( r \) es el radio del círculo.

- **Área de un círculo:**  
  $$
  A_{\text{círculo}} = \pi r^2
  $$

- **Perímetro de un rectángulo:**  
  $$
  P_{\text{rectángulo}} = 2(l + a)
  $$
  Donde \( l \) es el largo y \( a \) es el ancho.

- **Área de un rectángulo:**  
  $$
  A_{\text{rectángulo}} = l \cdot a
  $$



#### **Comparación de áreas:**
Supongamos que tenemos un potrero circular y un potrero rectangular con el mismo perímetro $P$.

1. **Para el círculo:**
   - El radio se calcula como:
     $$
     r = \frac{P}{2\pi}
     $$
   - El área es:
     $$
     A_{\text{círculo}} = \pi \left(\frac{P}{2\pi}\right)^2 = \frac{P^2}{4\pi}
     $$

2. **Para el rectángulo:**
   - Supongamos un rectángulo cuadrado (el rectángulo con mayor área para un perímetro dado). En este caso, $l = a$.
   - El lado se calcula como:
     $$
     l = \frac{P}{4}
     $$
   - El área es:
     $$
     A_{\text{rectángulo}} = \left(\frac{P}{4}\right)^2 = \frac{P^2}{16}
     $$

3. **Comparación:**
   - El área del círculo es:
     $$
     A_{\text{círculo}} = \frac{P^2}{4\pi} \approx \frac{P^2}{12.57}
     $$
   - El área del rectángulo cuadrado es:
     $$
     A_{\text{rectángulo}} = \frac{P^2}{16}
     $$

   Claramente, $\frac{P^2}{12.57} > \frac{P^2}{16}$, lo que significa que el círculo tiene un área mayor que el rectángulo para el mismo perímetro.

---



### **2. Implicaciones Prácticas**



#### **Mayor área útil:**
- Un potrero circular proporciona más espacio para el pastoreo que un potrero rectangular con el mismo perímetro. Esto es especialmente útil en regiones donde el terreno es limitado o costoso.

#### **Menor longitud de cerco:**
- Para un área dada, un potrero circular requiere menos longitud de cerco que un potrero rectangular. Esto reduce los costos de construcción y mantenimiento del cerco.

#### **Distribución uniforme:**
- La forma circular permite una distribución más uniforme de los recursos (agua, alimento) desde un punto central, lo que reduce la competencia entre los animales y mejora su bienestar.

---



### **3. Ejemplo Numérico**

Supongamos un perímetro $P = 100$ metros.

1. **Potrero circular:**
   - Radio:
     $$
     r = \frac{100}{2\pi} \approx 15.92 \text{ metros}
     $$
   - Área:
     $$
     A_{\text{círculo}} = \pi (15.92)^2 \approx 796 \text{ m}^2
     $$

2. **Potrero rectangular cuadrado:**
   - Lado:
     $$
     l = \frac{100}{4} = 25 \text{ metros}
     $$
   - Área:
     $$
     A_{\text{rectángulo}} = 25 \times 25 = 625 \text{ m}^2
     $$



3. **Comparación:**
   - El potrero circular tiene un área de **796 m²**, mientras que el potrero rectangular tiene un área de **625 m²**. Esto significa que el potrero circular ofrece un **27% más de área útil** para el mismo perímetro.

---



### **4. Conclusión**

La ventaja geométrica de los potreros circulares radica en su capacidad para maximizar el área útil con un perímetro dado. Esto no solo optimiza el uso del terreno, sino que también reduce costos de construcción y mejora la eficiencia en la distribución de recursos. En aplicaciones agropecuarias, esta propiedad hace que los potreros circulares sean una opción más eficiente y sostenible que los rectangulares.

# Recurso para trazado de una circunferencia dado en centro y su radio

````python
def trazar_circunferencia_puntos_cardinales(centro, radio): 
    """_summary_

    Args:
        centro (tuple): _description_
        radio (float): _description_
    """
    import matplotlib.pyplot as plt
    import numpy as np

    h,k = 5,7
    r = 7

    theta = np.linspace(0, 2 * np.pi, 100)
    x = h + r * np.cos(theta) # 
    y = k + r * np.sin(theta)

    plt.figure(figsize=(8, 8))
    plt.plot(x, y, label='Circunferencia')
    # Trazado del centro
    plt.scatter(h, k, color='red', label=f'Centro ({h}, {k}) y radio {r}')
    plt.axis('equal')
    plt.legend(loc = 'upper right')
    plt.title(f'Circunferencia $(x - {h})^2 + (y - {k})^2 = ({r})^{2}$')
    plt.grid(alpha = 0.4) 
    # Trazado de los puntos de la circunferencia correspondientes a 
    # los puntos cardinales  
    norte = (h, k+r)
    sur = (h, k-r)
    este = (h+r, k)
    oeste = (h-r, k)
    abscisas_puntos_cardinales, ordenadas_puntos_cardinales = zip(norte, sur, este, oeste)
    plt.scatter(abscisas_puntos_cardinales, ordenadas_puntos_cardinales, color='purple')
    plt.show()

````

````python
trazar_circunferencia_puntos_cardinales((5,7), radio = 7)
````

````python
# Diseño de un sistema de riego circular para una parcela en Caucasia, Antioquia
# Autor: Marco Julio
# Objetivo: Representar gráficamente y simular la cobertura de un sistema de riego circular

import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la parcela y del sistema de riego
radio_riego = 15  # metros, radio de cobertura del aspersor
largo_parcela = 60  # metros
ancho_parcela = 40  # metros
distancia_entre_aspersores = 2 * radio_riego * 0.9  # traslape del 10% entre coberturas

# Crear grilla de puntos donde colocar los aspersores
x_positions = np.arange(0, largo_parcela, distancia_entre_aspersores)
y_positions = np.arange(0, ancho_parcela, distancia_entre_aspersores)
aspersores = [(x, y) for x in x_positions for y in y_positions]

# Visualización del sistema de riego
fig, ax = plt.subplots(figsize=(10, 7))

# Dibujar la parcela
ax.set_xlim(0, largo_parcela)
ax.set_ylim(0, ancho_parcela)
ax.set_aspect('equal')
ax.set_title('Diseño de sistema de riego circular en Caucasia, Antioquia')
ax.set_xlabel('Metros (largo)')
ax.set_ylabel('Metros (ancho)')

# Dibujar los aspersores y su área de cobertura
for aspersor in aspersores:
    circulo = plt.Circle(aspersor, radio_riego, color='blue', alpha=0.2)
    ax.add_patch(circulo)
    ax.plot(aspersor[0], aspersor[1], 'ko')  # punto negro para el aspersor

plt.grid(True)
plt.show()

# Imprimir resumen del diseño
print(f"Número total de aspersores: {len(aspersores)}")
print(f"Cobertura de cada aspersor: {np.pi * radio_riego**2:.2f} m²")
print(f"Cobertura total aproximada: {len(aspersores) * np.pi * radio_riego**2:.2f} m²")
print(f"Área de la parcela: {largo_parcela * ancho_parcela} m²")
````

La función toma una cadena con parte de tu nombre, busca la cédula en un archivo Excel, extrae los últimos dígitos no nulos para definir el centro $(h, k)$ y el radio $r$ de la circunferencia, verifica si el punto $(2, 3)$ pertenece a la circunferencia y finalmente devuelve 4 puntos de la circunferencia.

````python
import pandas as pd
import math

def construir_circunferencia(nombre):
    # Ruta del archivo Excel
    ruta_archivo = r"C:\Users\marco\Documentos\docencia\groups_list\vectorial.xls"
    
    df = pd.read_csv(ruta_archivo, sep = ',')
    
    # Buscar la cédula correspondiente al nombre
    cedula = None
    for index, row in df.iterrows():
        if nombre.lower() in str(row.get("Nombre", "")).lower():
            cedula = row.get("Cedula")
            break
    
    if cedula is None:
        return "No se encontró una cédula asociada al nombre proporcionado."
    
    # Extraer los últimos dígitos no nulos de la cédula
    cedula_str = str(cedula).replace("0", "")  # Eliminar ceros
    ultimos_digitos = cedula_str[-3:]  # Tomar los últimos 3 dígitos no nulos
    
    # Asignar h, k y r
    r = int(ultimos_digitos[0])
    k = int(ultimos_digitos[1])
    h = int(ultimos_digitos[2])
    
    # Ecuación canónica de la circunferencia
    ecuacion = f"(x - {h})^2 + (y - {k})^2 = {r}^2"
    
    # Verificar si el punto (2, 3) pertenece a la circunferencia
    distancia_al_centro = math.sqrt((2 - h)**2 + (3 - k)**2)
    pertenece = distancia_al_centro == r
    
    # Calcular 4 puntos de la circunferencia
    puntos = [
        (h + r, k),  # Derecha
        (h - r, k),  # Izquierda
        (h, k + r),  # Arriba
        (h, k - r)   # Abajo
    ]
    
    # Resultados
    resultados = {
        "Ecuación de la circunferencia": ecuacion,
        "Centro (h, k)": (h, k),
        "Radio r": r,
        "¿El punto (2, 3) pertenece?": pertenece,
        "4 puntos de la circunferencia": puntos
    }
    
    return resultados
````

````python


# Ejemplo de uso
nombre = "amaris"  # Cambia esto por la parte de tu nombre
resultados = construir_circunferencia(nombre)
resultados

````

### Explicación del código:
1. **Lectura del archivo Excel**: Se utiliza `pandas` para leer el archivo Excel y buscar la cédula asociada al nombre proporcionado.
2. **Extracción de dígitos**: Se eliminan los ceros de la cédula y se toman los últimos 3 dígitos no nulos para definir $h$, $k$ y $r$.
3. **Ecuación de la circunferencia**: Se construye la ecuación canónica de la circunferencia.
4. **Verificación del punto \((2, 3)\)**: Se calcula la distancia del punto al centro y se compara con el radio.
5. **Puntos de la circunferencia**: Se calculan 4 puntos en la circunferencia (derecha, izquierda, arriba y abajo).

---



### Ejemplo de salida:
Si la cédula asociada al nombre es `1230456`, los últimos dígitos no nulos serían `456`, por lo que:
- $h = 4$, $k = 5$, $r = 6$.
- La ecuación de la circunferencia sería: $(x - 4)^2 + (y - 5)^2 = 36$.
- El punto $(2, 3)$ se verifica si pertenece.
- Los 4 puntos de la circunferencia serían: \((10, 5)\), \((-2, 5)\), \((4, 11)\), $(4, -1)$.
